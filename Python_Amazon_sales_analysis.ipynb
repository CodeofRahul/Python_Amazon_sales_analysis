## **Exploratory Data analysis**

```python
df.columns
```

```python
ax = sns.countplot(df, x='Size')

for bars in ax.containers:
  ax.bar_label(bars)
```

- From above bar graph we can see that most of the people are buying M-size.

```python
S_Qty = df.groupby(['Size'])['Qty'].sum().reset_index().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size', y='Qty', data=S_Qty)
```

```python
sns.countplot(df, x='Courier Status', hue='Status')
```

- we can see in the bar graph , the majority orders are shipped through the courier.

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=20, edgecolor='black')
plt.xticks(rotation=90)
plt.show()
```

- In above bar graph can be seen that most of the buyers are buying "T-shirt".

```python
# checking B2B data by using pie chart
B2B_check = df['B2B'].value_counts()

# plot the pie chart
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')

# plt.axis('equal')
plt.show()
```

- From above chart we can see that 99.2% of buyers are retailers and rest 0.8% are B2B buyers.

```python
x_data = df['Category']
y_data = df['Size']

# plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Categroy')
plt.ylabel('Size')
plt.title('Size by Category')
plt.show()
```

```python
# top 10 State
top_10_state = df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x='ship-state', hue='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution by State')
plt.xticks(rotation=45)
plt.show()
```

- From above graph we can seen that most of the buyers are from Maharashtra state.

### **Conclusion**

- The data analysis reveals that the business has a significant customer base in Maharashtra state.

- mainly serves retailers.

- fulfills orders through Amazon, experiences high demand for T-shirts and sees M-Size as the preferred choice of Amazon buyers.
